# Data Manipulation using Pandas

```python
import pandas as pd
```

```python
#Converting a list into array using Numpy
my_list = [9,5,7,6,1,8,2,3,4]
my_array = np.array(my_list)
print(my_array)
```

```python
#Creation of Data Series using Pandas with the help of lists 
#Converting lists into Series using Pandas
my_series1 = pd.Series(data=my_list)
print(my_series1)
```

```python
#Converting arrays into series using Pandas
my_series2 = pd.Series(data= my_array)
print(my_series2)
```

```python
print(my_series2[7])
```

```python
my_list1=[3,2,4,1]
my_series3 = pd.Series(data=my_list1)
```

```python
my_labels = ['zero', 'first', 'second', 'third']
my_series3 = pd.Series(data=my_list1, index=my_labels)
print(my_series3)
```

```python
my_series3 = pd.Series(my_list1, my_labels)
print(my_series3)
```

```python
print(my_series3['second'])
```

```python
#Addition of two Series
my_series4 = pd.Series([5,1,8,6],['zero', 'first', 'second', 'third'])
print(my_series4)
```

```python
print(my_series4+my_series3)
```

```python
#Concatenation of two Series
df = pd.concat([my_series3,my_series4], axis=1, sort=False)
df
```
